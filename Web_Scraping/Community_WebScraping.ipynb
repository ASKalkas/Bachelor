# Collecting Links

```python
def extract_links(driver, links):
    # Find the <ul> with class 'compactFeedList'
    ul_element = driver.find_element(By.XPATH, "//ul[@class='compactFeedList']")
    li_elements = ul_element.find_elements(By.XPATH, "./li[contains(@class, 'compactFeedListItem')]")

    for li in li_elements:
        # Find all <a> tags within the <li>
        a_tags = li.find_elements(By.TAG_NAME, "a")
        for a_tag in a_tags:
            link = a_tag.get_attribute("href")
            if link and link not in links:
                links.append(link)
    return links

links = []
extract_links(driver, links)
```

```python
def save_links_to_file(links, filename="links.txt"):
    with open(filename, "w") as file:
        for link in links:
            file.write(link + "\n")
    print(f"Saved {len(links)} links to {filename}")

save_links_to_file(links)
```

# Going through all links

```python
# Open the file in read mode
with open('links.txt', 'r') as file:
    # Read each line and strip any newline characters
    links = [line.strip() for line in file]

# Print the list of links to verify
print(links)
```

```python
def extract_data_from_link(driver, link):
    driver.get(link)
    time.sleep(5)  # Wait for the page to load completely
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    # Extract question title
    title_div = soup.find("div", class_="cuf-body cuf-questionTitle forceChatterFeedBodyQuestionWithoutAnswer")
    if title_div:
        question_title = title_div.get_text(strip=True)
    else:
        question_title = ""
    
    # Extract question body
    body_div = soup.find("div", class_="feedBodyInner Desktop")
    if body_div:
        question_body = body_div.get_text(strip=True)
    else:
        question_body = ""
    
    # Extract comments
    comments = []
    li_comments = soup.find_all("li", class_="cuf-commentLi published")
    if li_comments:
        for li in li_comments:
            # Find all <p> tags within the comment
            p_tags = li.find_all("p")
            comment_text = " ".join([p.get_text(strip=True) for p in p_tags])
            if comment_text:
                comments.append(comment_text)
    else:
        # Alternative method if comments are structured differently
        comment_divs = soup.find_all("div", class_="cuf-commentBody")
        for comment_div in comment_divs:
            comment_text = comment_div.get_text(strip=True)
            if comment_text:
                comments.append(comment_text)

    # Return the data as a dictionary
    data = {
        'link': link,
        'question_title': question_title,
        'question_body': question_body,
        'comments': comments
    }
    return data
```

```python
all_data = []

def collect_all_data(driver, links):
    for idx, link in enumerate(links):
        print(f"Processing link {idx+1}/{len(links)}: {link}")
        data = extract_data_from_link(driver, link)
        all_data.append(data)
    return all_data
```

```python
def save_data_to_csv(data_list, filename="data.csv"):
    # Prepare data for DataFrame
    rows = []
    for data in data_list:
        # Comments may have multiple entries; join them into a single string
        comments_text = " ||| ".join(data['comments'])
        rows.append({
            'Link': data['link'],
            'Question Title': data['question_title'],
            'Question Body': data['question_body'],
            'Comments': comments_text
        })
    # Create DataFrame
    df = pd.DataFrame(rows)
    # Save to CSV
    df.to_csv(filename, index=False, encoding='utf-8')
    print(f"Saved data to {filename}")
```
